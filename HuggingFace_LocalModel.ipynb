```python
def retrive(doc,doc_id):
  print(f'Most Similar Document: \n{doc[doc_id]}')
  return doc[doc_id]
```

```python
def user_input(doc, doc_embed, llm, emb_model):
  query = input('Enter your question: ')
  query_emb = emb_model.embed_query(query)
  scores = cosine_similarity([query_emb],doc_embed)[0]
  doc_id = max(list(enumerate(scores)),key=lambda x:x[1])[0]
  retrive(doc,doc_id)
  prompt = f"""<|system|>
  You are a helpful assistant.
  <|context|>{doc[doc_id]}
  <|user|>
  {query}"""

  result = llm.invoke(prompt)
  print(f'Question: {query}\n')
  print(prompt)

  print(f'Answer: {(result)}')
```
